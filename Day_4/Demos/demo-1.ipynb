```python
N = 11*7
a = 18

# N = 5*3
# a = 13
```

```python
import math
math.gcd(a, N)
```

```python
import matplotlib.pyplot as plotter

z = list(range(N))
y = [a**z0 % N for z0 in z]

plotter.plot(z, y)
plotter.xlabel('z')
plotter.ylabel(f'{a}^z (mod {N})')
plotter.show()
```

```python
r = z[y[1:].index(1)+1]
print(f'r = {r}')
```

```python
if r % 2 == 0:
    x = (a**(r/2.)) % N
    print(f'x = {x}')
    if ((x + 1) % N) != 0:
        print(math.gcd((int(x)+1),  N), math.gcd((int(x)-1), N))
    else:
        print("x + 1 is 0 (mod N)")
else:
    print(f'r = {r} is odd')
```
